```python
# Declare query parameters
start_index = 0
max_results_per_page = 100
total_results = 1000
query = 'natural immunity over covid vaccine'
tweet_fields = 'tweet.fields=text,created_at'
max_results = 'max_results=100'
```

```python
end_date = date.today()
days = timedelta(5)
start_date = end_date - days

end_date_str = str(end_date)
start_date_str = str(start_date)
```

```python
master_json_list = []
output = []
```

```python
# Store url in variable
url = f'https://api.twitter.com/2/tweets/search/recent?query={query}&{tweet_fields}&{max_results}&start_time={start_date}T00:00:00Z&end_time={end_date}T00:00:00Z'
```

```python
# Make API call and check for status

while start_index < total_results:
    new_url = f'{url}&startIndex={start_index}'
    res = requests.get(url, headers = headers)
    print("PRINTING RES: " ,res)
    res.raise_for_status()
    start_index += max_results_per_page
    json_data = res.json()
    print("PRINTING JSON_DATA: " , json_data)
    master_json_list += json_data['data']
```

```python
for item in master_json_list:
    values_list = []
    values_list.append(item['created_at'])
    print("VALUES_LIST: ",values_list)
    values_list.append(item['text'])
    print("VALUES_LIST2: ",values_list)
    output.append(values_list)
```

```python
for item in output:
    print(str(item) + '\n')
```
